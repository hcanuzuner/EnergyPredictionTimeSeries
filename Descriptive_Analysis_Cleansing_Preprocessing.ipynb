```python
#Saving to new variables so that we can reload the previous versions if needed
#I didn't want to overwrite the old variables with the original verisons of the file
#Seems to work well when merging data, had some problems with plotting and this resolved it

df_building = building_metadata
df_train = train
df_weather_train = weather_train
```

```python
#inspired by https://www.kaggle.com/mikechentf/ashrae-eda-and-basic-modeling
# Convert string timestamp to Pandas datetime
df_train['timestamp'] = pd.to_datetime(df_train['timestamp'])
df_weather_train['timestamp'] = pd.to_datetime(df_weather_train['timestamp'])
```

```python
#Inspired by:
# https://www.kaggle.com/mikechentf/ashrae-eda-and-basic-modeling
df_train_merge = df_train.merge(df_building, on='building_id', how='left').merge(df_weather_train, on=['site_id', 'timestamp'], how='left')
```

```python
#Code to establish new columns, month, weekend, etc.

# inspired by https://www.kaggle.com/ryches/simple-lgbm-solution
#df_train_merge["timestamp"] = pd.to_datetime(df_train_merge["timestamp"])
#df_train_merge["hour"] = df_train_merge["timestamp"].dt.hour
#df_train_merge["day"] = df_train_merge["timestamp"].dt.day
# df_train_merge["weekend"] = df_train_merge["timestamp"].dt.weekday
# df_train_merge["weekend"] = df_train_merge["timestamp"].dt.weekday
#df_train_merge["month"] = df_train_merge["timestamp"].dt.month

# Replacing with Can's version:
df_train_merge["hour"] = df_train_merge.timestamp.dt.hour
df_train_merge["weekday"] = df_train_merge.timestamp.dt.weekday
df_train_merge["day"] = df_train_merge.timestamp.dt.day
df_train_merge["month"] = df_train_merge.timestamp.dt.month

# ...the timestamp has already been changed to datatime 2 cells above
# "weekday" maybe more descriptive than "weekend"?
```

```python
#Code from https://www.kaggle.com/zeynepkurban/ashrae-2/notebook
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))

    return df
```
